```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report

data = {
    'age': ['<=30', '<=30', '31…40', '>40', '>40', '>40', '31…40', '<=30', '<=30', '>40', '<=30', '31…40', '31…40', '>40'],
    'income': ['high', 'high', 'high', 'medium', 'low', 'low', 'low', 'medium', 'low', 'medium', 'medium', 'medium', 'high', 'medium'],
    'student': ['no', 'no', 'no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'no'],
    'credit_rating': ['fair', 'excellent', 'fair', 'fair', 'fair', 'excellent', 'excellent', 'fair', 'fair', 'fair', 'excellent', 'excellent', 'fair', 'excellent'],
    'buys_computer': ['no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'no', 'yes', 'yes', 'yes', 'yes', 'yes', 'no']
}

df = pd.DataFrame(data)
df
```

```python
# A1. Calculate the prior probability for each class
prior_probabilities = df['buys_computer'].value_counts(normalize=True)
print("Prior probabilities for each class:")
print(prior_probabilities)
print()

```

```python
# A2. Calculate the class conditional densities for various features & classes
# For simplicity, let's consider Gaussian (normal) distribution for continuous features
class_conditional_densities = df.groupby('buys_computer').apply(lambda x: x[['age', 'income', 'student', 'credit_rating']].apply(lambda col: col.value_counts(normalize=True)))
print("Class conditional densities:")
print(class_conditional_densities)
print()

```

```python
# A3. Test for independence between the 4 given features
# For simplicity, let's use the Chi-square test for independence
from scipy.stats import chi2_contingency

# Convert categorical variables to numerical labels
df_numerical = df.apply(lambda x: pd.factorize(x)[0])

# Perform the Chi-square test for independence
chi2, p, _, _ = chi2_contingency(pd.crosstab(df_numerical['buys_computer'], df_numerical[['age', 'income', 'student', 'credit_rating']]))
print("Chi-square test for independence p-value:", p)
print()

```
